```python
"""Download Japanese Anime Speech Dataset"""
from datasets import load_dataset

ds = load_dataset("joujiboi/japanese-anime-speech")
```

```python
# dsにダウンロードしたデータセットをローカル保存する
ds.save_to_disk("japanese-anime-speech-dataset")
```

```python
"""Download Japanese Anime Speech Dataset and save audio files locally"""
import os
from pathlib import Path
import soundfile as sf
import csv

# 保存先ディレクトリを作成
output_dir = Path("./data/japanese-anime-speech")
audio_dir = output_dir / "audio"
audio_dir.mkdir(parents=True, exist_ok=True)

print(f"保存先: {output_dir}")

# データセットの構造を確認
print("\nデータセットの構造:")
print(ds)
print("\n最初のデータの例:")
print(ds["train"][0].keys())

# 音声ファイルとメタデータを保存
metadata = []
for i, example in enumerate(ds["train"]):
    # 音声データを保存
    if "audio" in example:
        audio_data = example["audio"]
        
        # 音声ファイルを保存
        audio_path = audio_dir / f"{i:05d}.wav"
        sf.write(
            str(audio_path),
            audio_data["array"],  # 音声データの配列
            audio_data["sampling_rate"],  # サンプリングレート
        )
        
        # メタデータを記録
        metadata.append({
            "id": i,
            "audio_path": str(audio_path),
            "text": example.get("text", ""),  # テキストがあれば
            "speaker": example.get("speaker", ""),  # 話者情報があれば
            "sampling_rate": audio_data["sampling_rate"],
        })
    
    if (i + 1) % 100 == 0:
        print(f"{i + 1} 件保存完了...")

# メタデータをCSVとして保存
metadata_path = output_dir / "metadata.csv"
with open(metadata_path, "w", newline="", encoding="utf-8") as f:
    if metadata:
        writer = csv.DictWriter(f, fieldnames=metadata[0].keys())
        writer.writeheader()
        writer.writerows(metadata)

print(f"\n✅ 保存完了！")
print(f"📁 音声ファイル: {audio_dir}")
print(f"📄 メタデータ: {metadata_path}")
print(f"📊 合計: {len(metadata)} 件")
```
